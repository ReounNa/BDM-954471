### 2. Sub Plot

plot หลายๆรูป ในแกนเดียวกัน

```python
plt.figure(1, figsize=(8,15))    # first create 1 figure,# the figsize argument is a tuple of (width, height)
# สร้างกระดาษขนาด 8*15

plt.subplot(3,1,1)               # numrows=1, numcols=3, fignum=1
# plt รูป โดยมีขนาด 3แถว 1คอลลัม รูปที่1
plt.xlabel('x')
plt.ylabel('y')
plt.title(r'$f(x) = x$')
#ชื่อหัวตาราง rข้างหน้าเหมือนเป็นตัวฟังชันสมาการแปลงtext
plt.plot(y1,'r+')        # plot the y1 sequence
# r คือสีแดง + คือเครื่องหมาย

plt.subplot(3,1,2)               # numrows=1, numcols=3, fignum=2
# plt รูป โดยมีขนาด 3แถว 1คอลลัม รูปที่2
plt.xlabel('x')
plt.ylabel('y')
plt.title(r'$f(x) = x^2$')       # scientific "TeX" notation for superscript
#plt.plot(y2, color='darkorange') # plot the y2 sequence
plt.plot(y2, 'b*')

plt.subplot(3,1,3)               # numrows=1, numcols=3, fignum=3
# plt รูป โดยมีขนาด 3แถว 1คอลลัม รูปที่3
plt.xlabel('x')
plt.ylabel('y')
plt.title(r'$f(x) = x^3$')
plt.plot(y3, color='green')      # plot the y3 sequence

plt.show()
```

```python
#จารใช้บ่อยมาก เพราะสั่งให้แต่ละกราฟต่างกันได้
import numpy as np
def tab2float(text): 
    return np.array([float(item) for item in text.rstrip().split("\t")], dtype=float) #จารเขียนฟังชั่นเอง ก็อบจาก excelมาแล้วจะแปลงให้

ETavg = tab2float("0.815	0.631	0.798	0.832	0.887")
LRavg = tab2float("0.756	0.515	0.744	0.768	0.818")
MLPavg = tab2float("0.788	0.579	0.779	0.797	0.865")
RFavg = tab2float("0.815	0.630	0.813	0.817	0.898")
SVMavg = tab2float("0.807	0.614	0.815	0.799	0.893")

ETtest = tab2float("0.808	0.618	0.775	0.841	0.889")
LRtest = tab2float("0.752	0.509	0.716	0.788	0.808")
MLPtest = tab2float("0.797	0.598	0.753	0.841	0.871")
RFtest = tab2float("0.814	0.628	0.792	0.835	0.904")
SVMtest = tab2float("0.805	0.610	0.783	0.826	0.899")
```

```python
#พวกนี้ออกสอบ
import os
labels = ['ACC', 'MCC', 'Sn', 'Sp', 'AUC']

plt.rcParams.update({'font.size': 14})
#สามารถsetได้หลายค่านอกจากขนาดอีกษร
x = np.arange(len(labels))  # the label locations
width = 0.16  # the width of the bars
plt.figure(figsize=(2*8.27,11.69/2))

ax1 = plt.subplot2grid((1,2), (0,0))
# subplot นี้เรียกว่า subplot2grid จะกำหนด ตำแหน่งที่วางรูปได้
# (1,2) คือสร้างภาพที่อยู่ในกริดมี 1แถุว 2 ภาพ
# (0,0) อยู่ตำแหน่ง 0,0

ax1.bar(x - 2*width, ETavg, width, label='ET') #สร้างบาร์ใน ตน. นั้นๆ # x เราตั้งไว้ให้มันคือ ตน. ตรงกลาง
ax1.bar(x - width, LRavg, width, label='LR')
ax1.bar(x, MLPavg, width, label='MLP')
ax1.bar(x + width, RFavg, width, label='RF')
ax1.bar(x + 2*width, SVMavg, width, label='SVM')# สี bar จะ auto ถ้าไม่ตั้ง
ax1.text(-0.1, 1.0, 'A', transform=ax1.transAxes,  fontsize=22, weight='bold') # ให้ตัว A ไปปรากฏที่ ตน. ที่ตั้งไว้
# Add some text for labels, title and custom x-axis tick labels, etc.
ax1.set_ylim([0,1.05]) #น่าจะ ค่าแนวตั้งเริ่ม0-1.05
ax1.set_ylabel('Average Performances', fontsize=16) # ตัวหนังสือข้างกราฟ
ax1.set_xticks(x) #น่าจะ ตน. การวางชื่อใ่ต้บาร์
ax1.set_xticklabels(labels, fontsize=16, rotation=45) # ตัวหนังสืชื่อใต้บาร์

ax2 = plt.subplot2grid((1,2), (0,1))
ax2.bar(x - 2*width, ETtest, width, label='ET')
ax2.bar(x - width, LRtest, width, label='LR')
ax2.bar(x, MLPtest, width, label='MLP')
ax2.bar(x + width, RFtest, width, label='RF')
ax2.bar(x + 2*width, SVMtest, width, label='SVM')
ax2.text(-0.1, 1.0, 'B', transform=ax2.transAxes,  fontsize=22, weight='bold')
# Add some text for labels, title and custom x-axis tick labels, etc.
ax2.set_ylim([0,1.05])
ax2.set_ylabel('Average Performances', fontsize=16)
ax2.set_xticks(x)
ax2.set_xticklabels(labels, fontsize=16, rotation=45)
ax2.legend(ncol=3,handleheight=1, labelspacing=0.05, markerscale = 0.5, fontsize=14) #ใส่คำอธิบายข้างบนว่าอะไรสีอะไร

plt.tight_layout(pad=0.8, w_pad=3, h_pad=1.0) 
#สั่งให้กราฟแต่ละอัน ห่างกัน
plt.savefig('%s/Figure1.pdf'%(os.getcwd()), format='pdf', dpi=450)
plt.show()
```

### 3. Graph Type

```python
import matplotlib.pyplot as plt

fig = plt.figure(1, figsize=(15,6))

# bar chart example
accounts = ['fabiyamada', 'jedigeiss', 'scipio', 'stoodkev']
steem = [1100, 500, 1200, 2300]

plt.subplot(1,2,1)
plt.title('Bar chart')
plt.ylabel('Steem')
plt.bar(accounts, steem) # แกน (x,y)

# pie chart example
# pie chart บอกค่าความเป็นส่วนร่วมบราๆ
labels = ['fabiyamada', 'jedigeiss', 'scipio', 'stoodkev']
perc = [steem[0]/sum(steem), # calculate percentages from the amount of steem
        steem[1]/sum(steem), 
        steem[2]/sum(steem), 
        steem[3]/sum(steem)]

explode = (0, 0, 0.5, 0)     # "explode" the 3rd slice ('Scipio') เอาอันที่3ระเบิดแยกออกมา เท่ๆ

plt.subplot(1,2,2)
plt.title('Pie chart')
plt.axis('equal')            # Make sure the pie is drawn as a circle, X and Y axes equally sized
plt.pie(perc, labels=labels, autopct='%1.2f%%', # .2f คือกำหนดให้เลขมี2ตน.
        startangle=0, explode=explode)

plt.show()
```
